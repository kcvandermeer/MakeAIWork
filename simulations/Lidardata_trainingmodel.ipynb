```python
import pandas as pd
# print(pd.__version__)
import numpy as np

# make numpy values easier to read.
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
keras = tf.keras

from tensorflow.keras.models import Model
```

```python
lidar_data = pd.read_csv(
    "C:/Users/kcvan/MakeAIWork/KEVINPROJECT1/CSVfiles/04_L_300922 - 04_L_300922.csv",
    names=["Lidar1", "Lidar2", "Lidar3", "Lidar4", "Lidar5", "Lidar6", "Lidar7", "Lidar8", "Lidar9", "Lidar10",
           "Lidar11", "Lidar12", "Lidar13", "Lidar14", "Lidar15", "Lidar16", "SteeringAngle"])

lidar_data.head()  
```

```python
# print(lidar_data)
```

```python
lidar_sensors = lidar_data.copy()
lidar_angle = lidar_sensors.pop('SteeringAngle')
lidar_sensors.head()
```

```python
#print(lidar_sensors)
```

```python
#lidar_angle.head()
```

```python
lidar_sensors = np.array(lidar_sensors)
lidar_sensors
```

```python
normalize = keras.layers.Normalization()
normalize.adapt(lidar_sensors)
```

```python
norm_lidar_model_01 = tf.keras.Sequential([
    normalize,
    keras.Input(shape=(16, )),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(1)
])

norm_lidar_model_01.compile(
    loss = tf.keras.losses.MeanSquaredError(), 
    metrics = ['accuracy'],
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
)
```

```python
norm_lidar_model_01.fit(lidar_sensors, lidar_angle, epochs=500)
```

```python
val_lidar_data = pd.read_csv(
    "C:/Users/kcvan/MakeAIWork/KEVINPROJECT1/CSVfiles/validation_lidar_01.csv",
    names=["Lidar1", "Lidar2", "Lidar3", "Lidar4", "Lidar5", "Lidar6", "Lidar7", "Lidar8", "Lidar9", "Lidar10",
           "Lidar11", "Lidar12", "Lidar13", "Lidar14", "Lidar15", "Lidar16", "SteeringAngle"])

val_lidar_sensors = val_lidar_data.copy()
val_lidar_angle = val_lidar_sensors.pop('SteeringAngle')

val_lidar_sensors = np.array(val_lidar_sensors)
val_lidar_sensors
```

```python
norm_lidar_model_01.evaluate(val_lidar_sensors,  val_lidar_angle, verbose=2)
```

```python
norm_lidar_model_01.predict(val_lidar_sensors)
```
